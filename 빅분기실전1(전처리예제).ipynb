```python
#데이터 불러오기
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/데이터 셋/P210203.csv")
print(df)
```

```python
#이상치 구하기 (이상치 기준은 평균으로부터 1.5표준편차 만큼 떨어진 값)
upper=df['charges'].mean()+1.5*np.std(df['charges'])
lower=df['charges'].mean()-1.5*np.std(df['charges'])

#이상치들의 합 구하기
result = df[(df['charges']>=upper) | (df['charges']<=lower)]['charges'].sum()

#정수부분만출력
print(int(result))
```

#작업형 2유형

```python
import pandas as pd

train=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/데이터 셋/P210204-01.csv")
test=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/데이터 셋/P210204-02.csv")

print(train.info())
print(test.info())
```

```python
#데이터 전처리

#ID저장 및 제거
train_ID=train.iloc[:,0]
test_ID=test.iloc[:,0]
#print(train_ID)
#print(test_ID)
train=train.drop('ID',axis=1)
test=test.drop('ID',axis=1)
#print(train.info())
#print(test.info())
```

```python
#라벨 인코딩
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
train['Warehouse_block']=le.fit_transform(train['Warehouse_block'])
train['Mode_of_Shipment']=le.fit_transform(train['Mode_of_Shipment'])
train['Product_importance']=le.fit_transform(train['Product_importance'])
train['Gender']=le.fit_transform(train['Gender'])

le=LabelEncoder()
test['Warehouse_block']=le.fit_transform(test['Warehouse_block'])
test['Mode_of_Shipment']=le.fit_transform(test['Mode_of_Shipment'])
test['Product_importance']=le.fit_transform(test['Product_importance'])
test['Gender']=le.fit_transform(test['Gender'])
```

```python
print(train.info())
print(test.info())
```

```python
#트레이닝 데이터 분리
from sklearn.model_selection import train_test_split

X=train.drop(columns='Reached.on.Time_Y.N')
Y=train['Reached.on.Time_Y.N']

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2024)
```

```python
#모델링 및 학습
from sklearn.ensemble import RandomForestClassifier

rfr = RandomForestClassifier(random_state=2024)
rfr.fit(X_train,Y_train)
pred=rfr.predict(X_test)
```

```python
#모델 평가

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score

print(confusion_matrix(Y_test,pred))
print(accuracy_score(Y_test,pred))
print(recall_score(Y_test,pred))
print(precision_score(Y_test,pred))
print(f1_score(Y_test,pred))
```

```python
#최종 결과 예측
pred2=rfr.predict_proba(test)
print(pred2)
pd.DataFrame({'ID':test_ID,'Reached.on.Time_Y.N':pred2[:,1]}).to_csv("result.csv",index=False)

data2=pd.read_csv("result.csv")
data2
```
